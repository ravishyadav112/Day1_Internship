```python
Country_mode = df['country'].mode()[0]  ## calculating the mode value of country 
```

```python
df['country'] = df['country'].fillna(Country_mode)  ## Filling missing values with Country_mode
```

```python
mapping = {'United States' : 'USA' , 'United Kingdom':'UK' , 'United Arab Emirates' : 'UAE' , 'South Africa' : 'SA'}
# df['country'] = df['country'].replace(mapping)

def standardise_country(cell):
    parts = [i.strip() for i in cell.split(',')] 
    
    abbr =[mapping.get(i,i)for i in parts]
    
    return ','.join(abbr)

df['country'] = df['country'].apply(standardise_country)
```
